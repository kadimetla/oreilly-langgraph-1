# Demonstration of an Agent from "Scratch"
1. Import ChatGPT via the OpenAI API and test it
2. Write a special function that allows the OpenAI API (chatgpt, gpt-4.1) to use Python functions to take actions requested by me
3. Test it for some silly simple tasks

```python
from openai import OpenAI
client = OpenAI()

def llm_call(prompt):
    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[{"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content
    
llm_call("Hi! Tell me 5 jokes about an AI instructor risking it all on the line for teaching agents")
```

```python
def tool_to_write_file(file_path: str, content: str) -> str:
    """
    Tool to write content to files
    """
    
    try:
        with open(file_path, "w") as f:
            f.write(content)
    except:
        return "Failed!"
        
    return "File was written!"

tool_to_write_file("./sample_file.txt", "Hi follks! I love pancakes")
```

```python
def llm_with_tool(prompt):
    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[{"role": "system", "content": """
                   When you receive a task that involves writing to files,
                   your output should be that call to the function that you 
                   have access to named: 'tool_to_write_file(file_path, content)'
                   
                   For example:
                   User: 'I want you to create a file named lucas.txt' containing the name 'Lucas'
                   AI: 'tool_to_write_file("./lucas.txt", "Lucas")'
                   """},
                  {"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content

output_tool_call = llm_with_tool("Create a file named best_pancake_toppings.txt containing top 5 toppings.")
```

```python
def execute_tool_call_from_llm(tool_call_output: str):
    try:
        return exec(tool_call_output)
    except:
        return "Failed execution"

execute_tool_call_from_llm(output_tool_call)
```
